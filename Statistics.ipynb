```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv(r"C:\Users\SIDDHARTH\OneDrive\Desktop\student_scores.csv")
mean=df['Maths'].mean()
mean=df['Science'].mean()
mean=df['English'].mean()
mean=df['SST'].mean()

mode=df['Maths'].mode()
mode=df['Science'].mode()
mode=df['English'].mode()
mode=df['SST'].mode()

median=df['Maths'].median()
median=df['Science'].median()
median=df['English'].median()
median=df['SST'].median()

print(mean)
print(mode)
print(median)

plt.subplot(1,2,1)
sns.histplot(df['Maths'],bins=10,kde=True,color='royalblue')
plt.title("Histogram of students'Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

subject_means = df.iloc[:, 1:].mean()
print(" Highest Scoring Subject:", subject_means.idxmax())
print(" Lowest Scoring Subject:", subject_means.idxmin())

df['Total'] = df.iloc[:, 1:].sum(axis=1)
df['Average'] = df.iloc[:, 1:].mean(axis=1)
def get_category(avg):
    if avg >= 85: return 'Excellent'
    elif avg >= 70: return 'Good'
    elif avg >= 50: return 'Average'
    else: return 'Needs Improvement'
df['Category'] = df['Average'].apply(get_category)
print("Student Performance:\n", df[['Student', 'Total', 'Average', 'Category']])

print("Subject Toppers:")
for subject in df.columns[1:5]:
    topper = df[df[subject] == df[subject].max()]
    name = topper['Student'].values[0]
    score = topper[subject].values[0]
    print(f"{subject}: {name} ({score})")

```
